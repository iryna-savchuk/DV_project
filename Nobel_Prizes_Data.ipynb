Import, from the the sorce 'https://nobelprize.readme.io/reference/getting-started' the Nobel Prize dataset

```python
#url_prizes = "http://api.nobelprize.org/v1/prize.csv"
url_laureate = "http://api.nobelprize.org/v1/laureate.csv?gender=All"
#url_country = "http://api.nobelprize.org/v1/country.csv"

#prize = requests.get(url_prizes).content
laureate = requests.get(url_laureate).content
#country = requests.get(url_country).content

#prize_df = pd.read_csv(io.StringIO(prize.decode('utf-8')))
df =  pd.read_csv(io.StringIO(laureate.decode('utf-8')))
#country_df =  pd.read_csv(io.StringIO(country.decode('utf-8')))
```

# 3. Data Exploration

```python
columns_name = df.columns.tolist()
columns_name
```

```python
# Missing values in 'born' it is an institution
df[df['born'].isna()]
```

```python
# Having a look at other organizations
df[df['gender']=='org'].tail()
```

```python
# Fill with the year it was formed 2013, we also can filter in gender org type
df['born'] = df['born'].fillna(value='2013')
```

```python
#Check all genders:
df['gender'].unique()
```

```python
# Convert 'born' and 'died' columns in datatime
df['born'] = df['born'].str[:4].astype('int64')
df['died'] = df['died'].str[:4].astype('int64')
```

```python
# Add a new column with the age when thay received the prize
df['prizeAge'] = df['year'] - df['born']
```

```python
df.info()
```

```python
#check for null values
df.isnull().sum()
```
