## 2.3. Merge

```python
# Add 'prizeAmount' and 'prizeAmountAdjusted' to the laureate_df dataframe
df_merged = pd.merge(
    laureate_df, 
    prize_amount_df[['id', 'awardYear', 'prizeAmount', 'prizeAmountAdjusted']], 
    left_on = ['id','year'],
    right_on = ['id', 'awardYear'],
    how = 'left').drop(columns = ['awardYear'])
    
df_merged.head()
```

```python
# Or even we can drop everu gender = 'org'
df_merged = df_merged.drop(df_merged[df_merged['gender'] == 'org'].index)
```

```python
df_merged['gender'].unique()
```

```python
# Convert 'born' and 'died' columns in datatime
df_merged['born'] = df_merged['born'].str[:4].astype('int64')
df_merged['died'] = df_merged['died'].str[:4].astype('int64')
```

```python
# Add a new column with the age when thay received the prize
df_merged['prizeAge'] = df_merged['year'] - df_merged['born']
```
