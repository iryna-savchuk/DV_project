```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import plotly.graph_objects as go


from wordcloud import WordCloud, STOPWORDS
from PIL import Image
```

## Loading Dataset

```python
df = pd.read_csv('data/merged.csv')
df.head(3)
```

## WordCloud 

Inspired by this recourse:
https://github.com/amueller/word_cloud

```python
text = df['motivation'].values 

stopwords = set(STOPWORDS)

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords).generate(str(text))

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# lower max_font_size
wordcloud = WordCloud(stopwords=stopwords, max_font_size=40, background_color="white").generate(str(text))
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
