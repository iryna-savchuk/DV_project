```python
import pandas as pd
import numpy as np

import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv('data/merged.csv')
```

Add a new column with 'count' with 1 to have the total of awards count

```python
df['count'] = 1
df['total'] = len(df['count'])
df.head(2)
```

```python
df.shape
```

```python
sunburst_df =  df.groupby(['total','category', 'gender'], as_index=False).size()
sunburst_df
```

```python
fig_test = px.sunburst(sunburst_df, path=['total', 'category', 'gender'], 
                       values='size')

fig_test.show()
```

```python
sunburst_df['laureate'] = sunburst_df['gender'].apply(lambda x: 'organisation' if x=='org' else 'person')
sunburst_df
```

```python
fig_test2 = px.sunburst(sunburst_df, 
                        path=['total', 'laureate', 'category'], 
                        values='size',
                        color = 'laureate',
                        color_discrete_map={'(?)':'orange', 'person':'gold', 'organisation':'darkblue'},
                        title="TWO laureate types - SIX award categories"
                        )
fig_test2.show()
```

```python
fig_test3 = px.sunburst(sunburst_df, 
                        path=['total', 'category', 'laureate'], 
                        values='size',
                        color = 'laureate',
                        color_discrete_map={'(?)':'orange', 'person':'gold', 'organisation':'darkblue'},
                        title="TWO laureate types - SIX award categories"
                        )
fig_test2.show()
```
