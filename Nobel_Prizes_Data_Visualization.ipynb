Add 'prizeAmount' and 'prizeAmountAdjusted' to the laureate_df dataframe

```python
# To overcome the duplicates we filter by 'id' and 'category'
df_merged = pd.merge(laureate_df, prize_amount_df[['id', 'category', 'prizeAmount', 'prizeAmountAdjusted']], on = ['id', 'category'], how = 'left')
df_merged.head()
```

```python
df_merged.shape
```

Import data set about countries

```python
country = pd.read_csv('country_info.csv')
country.head()
```

```python
country.shape
```

Import data set about country codes

```python
countrycode = pd.read_csv('country_code.csv', encoding='latin-1')
countrycode.head()
```

```python
countrycode.shape
```

Merge both country datasets

```python
country_df = pd.merge(country, countrycode[['2_code', '3_code']], left_on = 'country_code', right_on = '3_code', how = 'left').drop(columns = ['3_code'])
country_df.head()
```

```python
country_df.shape
```

```python
# add a new row with missing values
new_row = pd.DataFrame([['Undefined']*len(country_df.columns)], columns=country_df.columns)
country_df = pd.concat([country_df, new_row], ignore_index=True)
country_df.shape
```

```python
country_df.tail()
```

Merge df_merged with country_df

```python
df_merged['bornCountryCode'] = df_merged['bornCountryCode'].fillna(value='Undefined')
```

```python
df_final = pd.merge(df_merged, country_df[['2_code', 'region', 'income']], left_on = 'bornCountryCode', right_on = '2_code', how = 'left').drop(columns = ['2_code'])
df_final.head()
```

```python
df_final.shape
```

```python
# Missing values in 'born' it is an institution
df_final[df_final['born'].isna()]
```

```python
# We will drop it, or we can fill with the year it was formed 2013, we also can filter in gender org type
# df_merged = df_merged.drop(index=915)  # if we drop it
df_final['born'] = df_final['born'].fillna(value='2013')
```

```python
# convert born and died columns in datatime
df_final['born'] = df_final['born'].str[:4].astype('int64')
df_final['died'] = df_final['died'].str[:4].astype('int64')
```

```python
# Add a new column with the age when thay received the prize
df_final['prizeAge'] = df_final['year'] - df_final['born']
```

## Age

```python
x = df['prizeAge']
hist_data = dict(type='histogram', x=x, marker=dict(color='silver'))
layout = dict(title=dict(text='Ages Distribution'))
fig_1 = go.Figure(data=hist_data, layout=layout)
fig_1.show(renderer='iframe')
```

## Gender

```python
df['gender'].unique()
```

```python
gender_labels = df['gender'].value_counts()
gender_values = (gender_labels / gender_labels.sum()) * 100
unique_genders = df['gender'].unique()

gender_data = dict(type='pie',
                        labels=unique_genders,
                        values=gender_values,
                        textposition='inside', 
                        hole=0.6,
                        marker=dict(colors=['#0077c2', '#349de6', '#81cdf0'])
                        )

gender_layout = dict(title=dict(text='Prizes by Gender')
                  )

gender = go.Figure(data=[gender_data], layout=gender_layout)

gender.show(renderer='iframe')
```

## Category 

```python
category_labels = df['category'].value_counts()
category_values = (category_labels / category_labels.sum()) * 100
unique_category = df['category'].unique()

bar_category_data = dict(type='bar',
                        x=unique_category,
                        y=category_values,
                        marker=dict(color=['#0077c2', '#1083d3', '#2290e4', '#349de6', '#4aa9e9', '#5cb5eb', '#6fc1ed', '#81cdf0', '#93d9f2']))

bar_category_layout = dict(title=dict(text='Prizes by Category'), xaxis=dict(title='category'), yaxis=dict(title='Percentage')
                  )

bar_fig = go.Figure(data=[bar_category_data], layout=bar_category_layout)

bar_fig.show(renderer='iframe')
```

## Age by Gender

```python
# Filter data by gender
male_data = df.loc[df['gender'] == 'male', 'prizeAge']
female_data = df.loc[df['gender'] == 'female', 'prizeAge']
org_data = df.loc[df['gender'] == 'org', 'prizeAge']

# Create traces for each group
male_hist = go.Histogram(x=male_data, name='Male', marker=dict(color='blue'))
female_hist = go.Histogram(x=female_data, name='Female', marker=dict(color='red'))
org_hist = go.Histogram(x=org_data, name='org', marker=dict(color='green'))

# Create layout
layout = go.Layout(title=dict(text='Ages Distribution by Gender'))

# Create figure and plot
fig_1 = go.Figure(data=[male_hist, female_hist, org_hist], layout=layout)
fig_1.show(renderer='iframe')
```

## Category by gender

```python
# Filter data by gender
male_data = df.loc[df['gender'] == 'male', :]
female_data = df.loc[df['gender'] == 'female', :]

# Compute category values for each group
male_category_labels = male_data['category'].value_counts()
female_category_labels = female_data['category'].value_counts()

# Create traces for each group
male_bar = go.Bar(x=male_category_labels.index, y=male_category_labels.values, name='Male', marker=dict(color='blue'))
female_bar = go.Bar(x=female_category_labels.index, y=female_category_labels.values, name='Female', marker=dict(color='red'))

# Create layout
layout = go.Layout(title=dict(text='Prizes by Category'), xaxis=dict(title='category'), yaxis=dict(title='Count'))

# Create figure and plot
fig = go.Figure(data=[male_bar, female_bar], layout=layout)
fig.show(renderer='iframe')
```

```python
# Compute country counts and sort in descending order
country_counts = df['bornCountry'].value_counts().sort_values(ascending=False)

# Select top 10 countries
top_countries = country_counts.iloc[:10]

# Create bar trace
bar_trace = go.Bar(x=top_countries.index, y=top_countries.values, marker=dict(color=['orange'] + ['blue']*(len(top_countries)-1)))

# Create layout
layout = go.Layout(title=dict(text='Counts by Country of Birth'), xaxis=dict(title='Country'), yaxis=dict(title='Count'))

# Create figure and plot
fig = go.Figure(data=[bar_trace], layout=layout)
fig.show(renderer='iframe')
```

```python
# Compute region counts and sort in descending order
region_counts = df['region'].value_counts().sort_values(ascending=False)

# Select top 10 regions
top_regions = region_counts.iloc[:10]

# Create bar trace
bar_trace = go.Bar(x=top_regions.index, y=top_regions.values, marker=dict(color=['orange'] + ['blue']*(len(top_regions)-1)))

# Create layout
layout = go.Layout(title=dict(text='Counts by Regions of Birth'), xaxis=dict(title='Region'), yaxis=dict(title='Count'))

# Create figure and plot
fig = go.Figure(data=[bar_trace], layout=layout)
fig.show(renderer='iframe')
```

```python
# Compute income counts and sort in descending order
income_counts = df['income'].value_counts().sort_values(ascending=False)

# Select top 10 income
top_income = income_counts.iloc[:10]

# Create bar trace
bar_trace = go.Bar(x=top_income.index, y=top_income.values, marker=dict(color=['orange'] + ['blue']*(len(top_income)-1)))

# Create layout
layout = go.Layout(title=dict(text='Counts by the Income of Birth Country'), xaxis=dict(title='Income'), yaxis=dict(title='Count'))

# Create figure and plot
fig = go.Figure(data=[bar_trace], layout=layout)
fig.show(renderer='iframe')
```

```python
# Define colors for each category
category_colors = {'physics': 'blue', 'chemistry': 'orange', 'peace': 'green', 'medicine': 'yellow', 'literature': 'pink', 'economics': 'red'}

# Group the data by category and region, count the occurrences of each region within each category, and select the top 10 occurrences for each category
category_counts = df.groupby('category')['region'].value_counts().sort_values(ascending=False).groupby(level=0).head(10)

# Create a list of colors for the bars, based on the category of each region
bar_colors = [category_colors[cat] for cat, _ in category_counts.index]

# Create the bar trace
bar_trace = go.Bar(x=category_counts.index.get_level_values(1), y=category_counts.values, marker=dict(color=bar_colors))

# Create the layout
layout = go.Layout(title=dict(text='Counts by Regions of Birth and Category'), xaxis=dict(title='Region'), yaxis=dict(title='Count'))

# Create the figure and show the plot
fig = go.Figure(data=[bar_trace], layout=layout)
fig.show(renderer='iframe')
```
